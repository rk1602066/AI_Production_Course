```
df = pd.read_csv('Fortune_11 (1).csv')
```

```
df
```

```
df.fillna(0)
```

```
df.fillna(2)
```

```
df.fillna({'Name':'none','Profit':0,'Growth':0,'Industry':'none'})
```

```
#all missing values will be filled with the previous values of that column
df.fillna(method = 'ffill')
```

```
#all missing values will be filled with the previous values of that column
df.fillna(method = 'pad')
```

```
#all missing values will be filled with the backword values of that column
df.fillna(method = 'bfill')
```

```
#Must specify a fill 'value' or 'method'
df.fillna(axis=0)
```

```
df.fillna(method='ffill',axis=0)
```

```
#nan values will fill by previous column values
df.fillna(method='ffill',axis=1)
```

```
#nan values will fill by backword column values
df.fillna(method='bfill',axis=1)
```

```
#nan values will fill by 0 but only one value for each column because limit =1
df.fillna(0,limit=1)
```

```
df.fillna(0,limit=2)
```

```
df.fillna(method='ffill',limit=1)
```

```
df.fillna(method='ffill',limit=2)
```

```
#inplace=true ->does not create new dataframe. modify the original one
df.fillna(5,inplace=True)
df
```
